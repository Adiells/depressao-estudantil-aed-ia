```python
fig = make_subplots(rows=1, cols=2)

# Adicionando os gr√°ficos
fig.add_trace(go.Histogram(x=df['Academic Pressure']), row=1, col=1)
fig.add_trace(go.Histogram(x=df['Study Satisfaction']), row=1, col=2)


# Ajuste do layout
fig.update_layout(
    title='Subplots: Academic Pressure e Study Satisfaction',
    xaxis_title='Academic Pressure',
    yaxis_title='Contagem',
    xaxis2_title='Study Satisfaction',
    yaxis2_title='Contagem',
    showlegend=True)


fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Fun√ß√£o para adicionar histogramas automaticamente
def add_histograms(fig, df, column, row, col, colors={0: 'blue', 1: 'red'}, seq= None):
    for depression_val, color in colors.items():
        fig.add_trace(
            go.Histogram(
                x=df[df['Depression'] == depression_val][column],
                name=f'Com depress√£o' if depression_val == 1 else 'Sem depress√£o',
                marker=dict(color=color),
                showlegend=(col == 1)
            ),
            row=row, col=col
        )


# Criando o gr√°fico com subplots
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Press√£o acad√™mica / Depress√£o', 'Satisfa√ß√£o com estudos / Depress√£o')
)

# Adicionando histogramas usando a fun√ß√£o
add_histograms(fig, df, 'Academic Pressure', row=1, col=1)
add_histograms(fig, df, 'Study Satisfaction', row=1, col=2)

# Atualizando o layout do gr√°fico
fig.update_layout(
    title='Histograma - Satisfa√ß√£o e press√£o causada pelos estudos em rela√ß√£o ao √≠ndice de depress√£o',
    xaxis_title='Press√£o acad√™mica',
    yaxis_title='Contagem',
    xaxis2_title='Satisfa√ß√£o com os estudos',
    yaxis2_title='Contagem',
    showlegend=True)

fig.show()
```

Baseado nesses inshigts, vimos que a press√£o acad√™mica est√° ligada quase diretamente com os casos de depress√£o, se acentuando principalmente quando o n√≠vel de press√£o √© maior ou igual a 3, vamos guardar essa informa√ß√£o.

E vendo pela satisfa√ß√£o, tamb√©m percebemnos uma grande impacto, apesar de n√£o ser uma diferan√ßa brusca como na press√£o, comseguimos vizualizar que o n√∫mero de pessoas com n√≠vel 1 de satisfa√ß√£o com depress√£o, √© maior que o dobro de pessoas sem. E o mesmo se aplica ao n√≠vel 2.

## üß† An√°lise Inicial: Fatores que Influenciam a Depress√£o Acad√™mica

Com base nos dados e gr√°ficos analisados, foi poss√≠vel identificar **dois fatores que t√™m impacto direto nos n√≠veis de depress√£o entre estudantes**:

> ‚ö†Ô∏è **Diferente de fatores imut√°veis como idade**, os dois pontos abaixo s√£o elementos do contexto acad√™mico que **podem ser ajustados ou melhorados na vida real** ‚Äî seja por mudan√ßas institucionais ou pelas pr√≥prias rotinas dos alunos.

---

1. **Press√£o Acad√™mica**: A press√£o acad√™mica √© um fator que influencia diretamente os n√≠veis de depress√£o entre os estudantes. Quanto maior a press√£o, maior a incid√™ncia de depress√£o.

2. **Horas di√°ria de estudo/trabalho**: A quantidade de horas que os estudantes dedicam ao estudo ou trabalho tamb√©m est√° relacionada aos n√≠veis de depress√£o. Estudantes que passam mais tempo estudando ou trabalhando tendem a apresentar n√≠veis mais altos de depress√£o.

3. **Satisfa√ß√£o Acad√™mica**: A satisfa√ß√£o acad√™mica √© um fator que pode influenciar os n√≠veis de depress√£o. Estudantes que est√£o satisfeitos com sua vida acad√™mica tendem a apresentar n√≠veis mais baixos de depress√£o.

> üß™ Mais abaixo, ser√° feita uma simula√ß√£o considerando, por exemplo, uma situa√ß√£o onde **nenhum estudante ultrapassa 5 horas de estudo/trabalho por dia**, para analisar poss√≠veis impactos positivos.


```python
print('Valor real')
display(df['Depression'].value_counts(normalize=True))


print('Previs√µes')

filtro = df[df['Work/Study Hours'] <= 5]
print('Cortando quem estuda mais de 5 horas por dia')
display(filtro['Depression'].value_counts(normalize=True))


filtro = df[df['Study Satisfaction'] >= 3]
print('Cortando quem tem satisfa√ß√£o menor que 3')
display(filtro['Depression'].value_counts(normalize=True))

filtro = df[df['Academic Pressure'] <= 2]
print('Cortando quem tem pressao academica maior que 2')
display(filtro['Depression'].value_counts(normalize=True))

filtro = df[(df['Academic Pressure'] <= 2) & (df['Work/Study Hours'] <= 5)]
print('Exclue quem tem pressao academica maior que 2 e horas de estudo maior que 5')
display(filtro['Depression'].value_counts(normalize=True))

filtro = df[(df['Academic Pressure'] <= 2) & (df['Study Satisfaction'] >= 3)]
print('Exclue quem tem pressao academica maior que 2 e satisfa√ß√£o com estudos menor que 3')
display(filtro['Depression'].value_counts(normalize=True))

filtro = df[(df['Work/Study Hours'] <= 5) & (df['Study Satisfaction'] >= 3)]
print('Exclue quem tem horas de estudo maior que 5 e satisfa√ß√£o com estudos menor que 3')
display(filtro['Depression'].value_counts(normalize=True))

filtro = df[(df['Academic Pressure'] <= 2) & (df['Work/Study Hours'] <= 5) & (df['Study Satisfaction'] >= 3)]
print('Exclue quem tem pressao academica maior que 2, horas de estudo maior que 5 e satisfa√ß√£o com estudos menor que 3')
display(filtro['Depression'].value_counts(normalize=True))
```
